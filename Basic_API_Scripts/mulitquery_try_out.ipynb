```python
from multi_query import Multiquery
import pandas as pd
```

```python
myquery = Multiquery()
myquery.query_monthly_combination_result( search_string = "Wearable",
                                            search_nice_name= "Wearable",
    search_akronym_for_graph= "Wearable",
    norm_query= "ALL",
    start_month = "2000-01-01",
    end_month = "2022-10-01",
    show_rolling_average = True,
    twitter = True,
    window_rolling_average = 12)
```

```python
ctgov_data = pd.read_csv("Graphs/DBio/DBio_CTGOV.csv")
pubmed_data = pd.read_csv("Graphs/DBio/DBio_PubMed.csv")
twitter_data = pd.read_csv("Graphs/DBio/DBio_Twitter.csv")
```

```python
twitter_data["month"] = pd.to_datetime(
    twitter_data["month"], infer_datetime_format=True
)
twitter_data.groupby(twitter_data.month.dt.year)["tweet_count"].mean()
```

```python
ctgov_data["month"] = pd.to_datetime(
    ctgov_data["month"], infer_datetime_format=True
)

a = ctgov_data.groupby(ctgov_data.month.dt.year)["trial_count"].sum()
b = ctgov_data.groupby(ctgov_data.month.dt.year)["reference_total_trial_count"].sum()
a/b
```

```python
pubmed_data["month"] = pd.to_datetime(
    pubmed_data["month"], infer_datetime_format=True
)

a = pubmed_data.groupby(pubmed_data.month.dt.year)["pub_count"].sum()
b = pubmed_data.groupby(pubmed_data.month.dt.year)["norm_pub_count"].sum()
a/b
```
